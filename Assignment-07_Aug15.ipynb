3. **Filtering and Sorting:**
   - Filter the dataset to show only the rows where the `sepal_length` is greater than 5.0.
   - Sort the filtered dataset by `petal_length` in descending order.


```python
filtered_df = df[df['SepalLengthCm'] > 5.0]
print("\n Rows with Sepal Length > 5.0")
print(filtered_df.head())
```

```python
sorted_df = filtered_df.sort_values('PetalLengthCm', ascending=False)
print("\nFiltered Data Sorted by Petal Length in Reverse Order")
print(sorted_df.head())
```

5. **Creating New Columns:**
   - Create a new column called `sepal_petal_ratio` which is the ratio of `sepal_length` to `petal_length`.


```python
df['sepal_petal_ratio'] = df['SepalLengthCm'] / df['PetalLengthCm'] # i created new column sepal_petal_ratio

print("\nDataFrame with new column")
print(df[['SepalLengthCm', 'PetalLengthCm', 'sepal_petal_ratio']].head())
print("\n Stats for Sepal-Petal Ratio ")
print(df['sepal_petal_ratio'].describe())
```

6. **Data Visualization:**
   - **Histogram:**
     - Create a histogram to visualize the distribution of `sepal_length` values.
   - **Scatter Plot:**
     - Create a scatter plot with `sepal_length` on the x-axis and `petal_length` on the y-axis. Use different colors to represent different species.
   - **Box Plot:**
     - Create a box plot for the `sepal_width` values, grouped by `species`.

```python
import matplotlib.pyplot as roksh
import seaborn as sns

roksh.figure(figsize=(8, 5))
sns.histplot(data=df, x='SepalLengthCm', bins=15, kde=True)
roksh.title('Distribution of Sepal Length (cm)')
roksh.xlabel('Sepal Length (cm)')
roksh.ylabel('Frequency')
roksh.show()
```

```python
roksh.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='SepalLengthCm', y='PetalLengthCm', 
                hue='Species', palette='viridis', s=80)
roksh.title('Sepal Length vs Petal Length by Species')
roksh.xlabel('Sepal Length (cm)')
roksh.ylabel('Petal Length (cm)')
roksh.legend(title='Species')
roksh.show()
```

```python
roksh.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Species', y='SepalWidthCm', 
            palette='pastel', width=0.5)
roksh.title('Sepal Width Distribution by Species')
roksh.xlabel('Species')
roksh.ylabel('Sepal Width (cm)')
roksh.show()
```


7. **Correlation Matrix**
   - Calculate the correlation matrix for the numerical features and create a heatmap to visualize the correlations using `seaborn`.



```python
correlation_matrix = df.select_dtypes(include=['float64', 'int64']).corr() # calculating correlation matrix of only numerical columns

roksh.figure(figsize=(8, 6)) # creating heatmap
sns.heatmap(correlation_matrix, 
            annot=True,  
            cmap='coolwarm', 
            center=0, 
            fmt='.2f', 
            linewidths=0.5)

roksh.title('Correlation Matrix of Iris Dataset Features')
roksh.xticks 
roksh.yticks  
roksh.tight_layout()
roksh.show()
```
